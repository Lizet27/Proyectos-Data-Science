###**Datos nulos**


####**Limpieza de los datos nulos de la columna Embarked**

1. Muestra las observaciones que contengan los datos nulos en la columna Embarked.
2. Implementa una función para rellenar los datos nulos encontrados.

```python
#Respuesta 1:
df[df['Embarked'].isnull()]
```

```python
#Respuesta 2:
def fillEmbarked(df):
  data=df['Embarked'].mode()[0]
  df['Embarked'].fillna(data, inplace=True)
  return df
fillEmbarked(df)
```

```python
#Respuesta 2:
def fillAge(df):
  data=df['Age'].mode()[0]
  df['Age'].fillna(data, inplace=True)
  return df
fillAge(df)
```


##4.   **Visualización de los datos**


###**Gráfico número 1**
**Distribución porcentual de pasajeros según supervivencia**


```python
#Coloca tu código aquí:
from pandas.io.formats import style
dis_sur = df['Survived'].value_counts(normalize=True) * 100
sty= dis_sur.rename("Porcentaje").reset_index()
def format_percentage(x):
    return f"{x:.2f}%"
# Crear el gráfico circular
plt.figure(figsize=(6, 6))
plt.pie(dis_sur, labels=["No sobrevivió", "Sobrevivió"], autopct='%1.2f%%', colors=["red", "lightblue"])
plt.ylabel("Survived")
plt.title("Distribución porcentual de pasajeros según supervivencia")

# Mostrar el
plt.show()
```

###**Gráfico número 2**
**Distribución de pasajeros según supervivencia por sexo y clase**


```python
#Coloca tu código aquí
distr_pasa = df.groupby(['Sex', 'Pclass', 'Survived']).size().unstack().fillna(0)

# Crear un gráfico de barras apiladas
distr_pasa.plot(kind='bar', stacked=True, color=['lightcoral', 'lightgreen'], figsize=(10, 6))

# Etiquetas y título del gráfico
plt.title('Distribución de pasajeros según supervivencia, sexo y clase')
plt.xlabel('Sexo y Clase')
plt.ylabel('Cantidad de Pasajeros')

# Mostrar el gráfico
plt.legend(['No sobrevivió', 'Sobrevivió'])
plt.show()
```
